```python
import tfcoreml
```

```python
labels = [label[:-1] for label in open("./labels.txt").readlines()]
```

```python
tfcoreml.convert(
    class_labels=labels,
    tf_model_path="./graph_mobilenet.pb",
    image_input_names=['Placeholder:0'],
    mlmodel_path="./realmodel.mlmodel",
    output_feature_names=['final_result:0'],
    input_name_shape_dict = {'Placeholder:0' : [1, 224, 224, 3]},
    red_bias = -1,
    green_bias = -1,
    blue_bias = -1,
    image_scale = 2.0/255.0
)
```
